```python
vidcap = cv2.VideoCapture('/users/iamns45/Desktop/CV/CV-3/Que-1/Que3-1.mp4')
success,image = vidcap.read()
count = 0
while success:
    success,image = vidcap.read()
    if count%30==0 :
      gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
      cv2.imwrite("/users/iamns45/Desktop/CV/CV-3/Que-1/frame%d.jpg" % count, gray_image)     # save frame as JPEG file      
      print('Stored a frame in frame%d.jpg'%count, success)
    count += 1
```

```python
for i in range(0,240,30):
    gray = cv2.imread("/users/iamns45/Desktop/CV/CV-3/Que-1/frame%d.jpg"%i, 0)
    patch= cv2.imread("/users/iamns45/Desktop/CV/CV-3/Que-1/Cropped.jpg", 0)
    ssd = cv2.matchTemplate(gray, patch, cv2.TM_SQDIFF_NORMED) 
    point = np.where(ssd == ssd.min())
    y = point[0][0]
    x = point[1][0]
    w = len(patch[0])
    h = len(patch)
    plt.figure(figsize = (20,2))
    cv2.rectangle(gray, (x, y), (x+w, y+h), (255, 2, 1), 2)
    plt.axis('off')
    plt.imshow(gray,cmap='gray', vmin=0, vmax=255)
    plt.show()
```
